```python
# df = pd.read_csv('/Users/jornelasmunoz/Downloads/CodedAperture_CIFAR10_classification.csv')
df = pd.read_csv('/Users/jocelynornelas/Downloads/CodedAperture_MNIST_classification - MNIST.csv')
df.head()
direct_model = df[df['Model']== 'Direct']
recon_model = df[df['Model']== 'Reconstructed']
```

```python
# Set up the Seaborn style
sns.set(style="whitegrid", font="Times New Roman" )

# Create a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=direct_model['Trained Model'],y=direct_model['Accuracy'], data=direct_model, hue=direct_model['Tested Model'], palette='Blues')

# Customize the chart
plt.title('Accuracy of Direct Method on MNIST', fontsize=18)
plt.xlabel('Noise Level of Training Data', fontsize=20)
plt.ylabel('Accuracy (%)', fontsize=20)
plt.xticks(rotation=0, ha='right', fontsize=18)  # Rotate x-axis labels for better readability
plt.yticks(fontsize=14)
plt.legend(title='Tested Models', loc='lower right', fontsize='large')#,bbox_to_anchor=(0.20, -0.5), fontsize='large',ncol=2)
plt.ylim([0,100])

# plt.savefig('../figs/CIFAR_accuracy_direct_all_noise.png')
plt.savefig('../figs/MNIST_accuracy_direct_all_noise.png')
plt.show()
```

```python
# Set up the Seaborn style
sns.set(style="whitegrid", font="Times New Roman" )

# Create a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=recon_model['Trained Model'],y=recon_model['Accuracy'], data=recon_model, hue=df['Tested Model'], palette='Blues')

# Customize the chart
plt.title('Accuracy of Reconstructed Method on MNIST', fontsize=18)
plt.xlabel('Noise Level of Training Data', fontsize=20)
plt.ylabel('Accuracy (%)', fontsize=20)
plt.xticks(rotation=0, ha='right', fontsize=18)  # Rotate x-axis labels for better readability
plt.yticks(fontsize=14)
plt.legend(title='Tested Models', loc='lower right', fontsize='large')#,bbox_to_anchor=(0.20, -0.5), fontsize='large',ncol=2)
plt.ylim([0,100])

# plt.savefig('../figs/CIFAR_accuracy_recon_all_noise.png')
plt.savefig('../figs/MNIST_accuracy_recon_all_noise.png')

# Show the plot
plt.show()
```

```python
# Set up the Matplotlib figure
plt.figure(figsize=(12, 8))

# Iterate over unique trained models and superimpose bar plots
for model in df['Model'].unique():
    subset_df = df[df['Model'] == model]
    sns.barplot(x=subset_df['Trained Model'],y=subset_df['Accuracy'], data=subset_df, hue=df['Tested Model'], palette='Greens' if model == 'Direct' else 'tab20c', alpha = 1 if model == 'Direct' else .3)
    # plt.bar(subset_df['Tested Model'], subset_df['Accuracy'], label= model)

# Customize the chart
plt.title('Superimposed Bar Plot of Accuracy')
plt.xlabel('Tested Models')
plt.ylabel('Accuracy')
plt.legend(title='Trained Models', loc='lower left')
# plt.ylim([50,100])
# Show the plot
plt.show()
```

```python
from matplotlib.colors import LinearSegmentedColormap

# Create a colormap
original_cmap = plt.cm.tab20c  # You can use any other existing colormap

# Get the colormap colors
colors = original_cmap(np.linspace(0, 1, original_cmap.N))

# # Change the start color (index 0)
# new_start_color = [1, 0, 0, 1]  # RGBA values, you can modify as needed
# # colors[0] = new_start_color

# Create a new colormap
new_cmap = LinearSegmentedColormap.from_list('new_colormap', colors[8:], N=original_cmap.N-8)

# Plot using the new colormap
data = np.random.rand(10, 10)
plt.imshow(data, cmap=new_cmap)
plt.colorbar()

plt.show()
```

```python
import scipy.stats as stats

# Perform independent two-sample t-test
t_stat, p_value = stats.ttest_ind(list(direct_model.Accuracy), list(recon_model.Accuracy))

# Check if the p-value is less than your chosen significance level (e.g., 0.05)
if p_value < 0.05:
    print("The difference is statistically significant.")
else:
    print("The difference is not statistically significant.")
```
